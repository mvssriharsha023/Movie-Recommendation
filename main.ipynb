## Content based recommendation

```python
data.head()
```

```python
data['Combined'] = data['Title'] + ' ' + data['Director'] + ' ' + data['Actor1'] + ' ' + data['Actor2'] + ' ' + data['Actor3'] + ' ' + data['Genre'] + ' ' + data['Overview']
```

```python
data.head()
```

```python
data.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count_vectorizer = CountVectorizer(max_features=5790, stop_words='english')
```

```python
vector = count_vectorizer.fit_transform(data['Combined'].values.astype('U')).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend_movies(movie):
    index = data[data['Title'] == movie].index[0]
    recommended_movies = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])[1: 21]
    df = pd.DataFrame()
    for movie in recommended_movies:
        series = pd.Series(data.iloc[movie[0]])
        df = df.append(series, ignore_index=True)
    df = df.sort_values(by='Ratings', ascending=False).head(10)

    df = df.reset_index().drop(columns=['index', 'Combined'], axis=1)
    return df
```

```python
recommend_movies('Tenet')
```
